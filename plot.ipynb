```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eval_EM = np.load("./qalertlargegr64linearepo6re2/eval_EM.npy")
eval_f1 = np.load("./qalertlargegr64linearepo6re2/eval_f1.npy")
eval_tloss = np.load("./qalertlargegr64linearepo6re2/train_loss_curve.npy")/64

plt.figure(0)
plt.title('eval-EM&F1')
plt.xlabel('epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.ylim(0.8, 1.0) 
line1, = plt.plot(np.arange(eval_EM.shape[0]), eval_EM, color = 'red', linewidth = 3, label = 'eval EM')             
line2, = plt.plot(np.arange(eval_f1.shape[0]), eval_f1, color = 'orange', linewidth = 3, label = 'eval f1')

plt.legend(handles = [line1, line2], loc='upper right')
plt.show()
```

```python
plt.figure(1)
plt.title('eval-Loss')
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.grid(True)
#line4, = plt.plot(np.arange(eval_loss.shape[0]), eval_loss, color = 'yellow', linewidth = 3, label = 'eval loss')
line3, = plt.plot(np.arange(eval_tloss.shape[0]), eval_tloss, linewidth = 3, label = 'train loss')
plt.legend(handles = [line3], loc='upper right')
plt.show()
```
